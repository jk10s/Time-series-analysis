```python
excel= pd.read_csv('finanzaspunto.csv',sep=";")
excel
```

```python
excel.dtypes
```

```python
excel['Fecha'] = pd.to_datetime(excel['Fecha'], format= '%d/%m/%y')
```

```python
excel
```

```python
excel.dtypes
```

```python
plt.style.use("ggplot")
fig, ad = plt.subplots(figsize=(15,8))

```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(excel['Elon Mosk'], label = 'Elon Mosk',lw =2, color = "blue", path_effects =[pa.SimpleLineShadow(),pa.Normal()])
ax.plot(excel['Bill Gate'], label = 'Bill Gate',lw =2, color = "red", path_effects =[pa.SimpleLineShadow(),pa.Normal()])
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(excel['Elon Mosk'], label = 'Elon Mosk',lw =2, color = "blue", path_effects =[pa.SimpleLineShadow(),pa.Normal()])
ax.plot(excel['Bill Gate'], label = 'Bill Gate',lw =2, color = "red", path_effects =[pa.SimpleLineShadow(),pa.Normal()])
ax.legend(shadow =True, prop ={"family":"serif", "size":15})
ax.set_title("elon vs bill gate")
ax.set_xlabel('USd')
ax.set_ylabel("time")
```
